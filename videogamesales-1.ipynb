-Which games had the highest sales by name, region, and platform?

```python
video_games[['Name', 'NA_Sales', 'JP_Sales', 'EU_Sales', 'Other_Sales']].head(20)
```

```python
sales_NA = video_games.sort_values(by='NA_Sales', ascending=False)[['Name', 'NA_Sales', 'Platform', 'Publisher']].head(20)
sales_NA
```

```python
sales_EU = video_games.sort_values(by='EU_Sales', ascending=False)[['Name', 'EU_Sales', 'Platform', 'Publisher']].head(20)
sales_EU
```

```python
sales_JP = video_games.sort_values(by='JP_Sales', ascending=False)[['Name', 'JP_Sales', 'Platform', 'Publisher']].head(20)
sales_JP
```

Looking at the popularity by sales, in NA and EU, Wii Sports takes number one. When the Wii was released, in those regions, Wii Sports came with the console, which would make sense for the popularity of that particular game. Interestingly, Wii Sports was not bundled in Japan, which would answer why it wasn't number one in that region, or even top 20 for that matter. 

Additionally, we're able to see the highest sales by platform, and Nintendo easily has the highest sales. It could be for a number of reasons, but Nintendo has always marketed towards families and children, and as of recently, the 'casual gamer'. Especially for a game like Wii Sports, the entry skill level is very approachable. The Wii controller wasn't overly complicated to use, even if you've never used a console controller. When the Wii first launched, it was branded as a way to spend time with the family, and that continues to be how they market themselves. 

-Does the genre have a overwhelming popularity?

```python
game_genre = video_games.sort_values(by='Global_Sales', ascending=False)['Genre']
unique_genres = game_genre.drop_duplicates()
unique_genres
```

With Wii Sports coming in at number one for most regions, it makes sense that "Sports" is the top genre. Nintendo not only creates a lot of platform games, but created the genre itself. Nintendo also appeared to rank high consistently with sales, so it only makes sense that platform would also be ranked high.
